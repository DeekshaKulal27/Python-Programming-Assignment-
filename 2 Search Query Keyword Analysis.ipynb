```python
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation (if present)
import string
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split into words
words = query.split()

# Step 4: Count frequency
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Step 5: Filter words with frequency > 1
repeated_words = {word: count for word, count in word_count.items() if count > 1}

print(repeated_words)
```
